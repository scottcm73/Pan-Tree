```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv('Resources/products.csv')
```

```python
def random_price():
    random_price=np.random.choice(20, 1, p=[0.01, 0.01, 0.4, .3, 0.1, 0.05, 0.03, .02, .01, .01, .01, .01, .01, .005, .005, .005, .005, .005, .0025, .0025 ])
    price = random_price+.99
    return price

df["price"] = 0
df["price"] = df["price"].apply(lambda x: x+ float(str(random_price())[1:-1]))

df.head()
```

```python
df.to_csv("Resources/products_price.csv")
```
